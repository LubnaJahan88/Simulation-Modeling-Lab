```python
#LUBNA JAHAN LIPA
#ID: 0432310005101088

import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)

uniform_wait = np.random.uniform(2, 12, 3000)

normal_wait = np.random.normal(8, 2, 3000)

normal_wait[normal_wait < 0] = 0
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

#Uniform
axes[0].hist(uniform_wait, bins=40, color='seagreen', edgecolor='black', alpha=0.7)
axes[0].set_title('Uniform Distribution')
axes[0].set_xlabel('Wait Time')
axes[0].set_ylabel('Frequency')

#Normal
axes[1].hist(normal_wait, bins=40, color='royalblue', edgecolor='black', alpha=0.7)
axes[1].set_title('Normal Distribution')
axes[1].set_xlabel('Wait Time')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
print("For Regular Days:")
print("Mean:", np.mean(uniform_wait))
print("Median:", np.median(uniform_wait))
print("Standard Deviation:", np.std(uniform_wait))
print("Maximum:", np.max(uniform_wait))

print("\nFor Busy Days:")
print("Mean:", np.mean(normal_wait))
print("Median:", np.median(normal_wait))
print("Standard Deviation:", np.std(normal_wait))
print("Maximum:", np.max(normal_wait))
```
