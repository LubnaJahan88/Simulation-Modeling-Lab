```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#ASSIGNMENT-01

L = np.random.randint(10,30,(3, 3))
J = np.random.randint(60,80,(3,3))

print(L)
print(J)
```

```python
Add=L+J
Sub=L-J
Mul=np.dot(L,J)

print(Add)
print(Sub)
print(Mul)
```

```python
det_L = np.linalg.det(L)

if det_L!=0:
    inv_L = np.linalg.inv(L)
else:
    inv_L = "Matrix is singular, cannot compute inverse"

rank_L = np.linalg.matrix_rank(L)

print(det_L)
print(inv_L)
print(rank_L)
```

```python
#ASSIGNMENT-02

x = np.random.rand(15)
y = np.random.rand(15)

print(x)
print(y)
```

```python
index = np.arange(len(x))
print(index)

plt.plot(index, x, marker='o', linestyle='-', label="Vector X")
plt.plot(index, y, marker='s', linestyle='--', label="Vector Y")

plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Line Graph of Two Random Matrices X and Y")
plt.legend()
plt.grid(True)
plt.show()
```
